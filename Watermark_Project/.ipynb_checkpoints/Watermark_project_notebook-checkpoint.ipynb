```python
################# Watermark implementation on an image ################# 

from skimage.io import imread,imsave,imshow
import numpy as np
from matplotlib import pyplot as plt
from PIL import Image, ImageDraw, ImageFont
%matplotlib inline 

#Take for arguments: a text t and a font f
#Evaluate the size needed for the watermark to contain the text
def size_of_text(t, f):
    im_for_text_size = Image.new('RGBA', (1, 1), (0,0,0,255))
    d = ImageDraw.Draw(im_for_text_size)
    return d.textsize("{}".format(t), f)

#Take for arguments:  the wanted text, its font and optionally if the user want to save it
#Create a RGBA watermark image from scratch with the wanted text
def create_im_watermak(t, f, save = None):
    s = size_of_text(t, f)
    i = Image.new("RGBA", (s[0], s[1]), (0,0,0,255))
    d = ImageDraw.Draw(i)
    d.text((0, -35), t, font=f, fill=(255,255,255,255))
    if save:
        i.save('{}.png'.format(t))
    return i 

#Take for argument: an image 
#Transform the image to RGBA
def to_rgba(im):
    if (np.array(im).ndim == 2):
        transparency = np.ones(im.shape)*255
        im_rgba = np.stack((im, im, im, transparency), axis=2).astype('uint8')
    elif (np.array(im).ndim == 3 and np.array(im).shape[2] == 3):
        r = im[:, :, 0]
        g = im[:, :, 1]
        b = im[:, :, 2]
        transparency = np.ones(r.shape)*255
        im_rgba = np.stack((r, g, b, transparency), axis=2).astype('uint8')
    else:
        im_rgba = im
    return im_rgba

#Take for arguments: the image and the watermark 
#Resize the watermark according to the image dimensions
def resize_watermark(im, wm):
    if (type(wm) != Image.Image):
        wm = Image.fromarray(wm)
    if (im.shape[0] <= im.shape[1]):
        wm = wm.resize((np.array(wm).shape[1]*im.shape[0]//(10*np.array(wm).shape[0]), im.shape[0]//10), Image.NEAREST)
    else:
        wm = wm.resize((im.shape[1]//5, np.array(wm).shape[0]*im.shape[1]//(5*np.array(wm).shape[1])), Image.NEAREST)
    return  np.array(wm)

#Take for arguments: the image, the watermark and the wanted position
#For the position it can be coordinates (list or tuple) or one of the 10 predetermined positions (str)
#Predetermined positions: l = left; m = middle ; r = right; a = ascendent ; d = descendent
#Return a tuple (y, x) of the wanted position
def convert_to_coordinates(im, wm, p):
    if (type(p) != str):
        pos = p
    elif (p == 'la'):
        pos = (5, 5)
    elif (p == 'lm'):
        pos = ((im.shape[0]//2)-wm.shape[0]//2, 5)
    elif (p == 'ld'):
        pos = ((im.shape[0]-wm.shape[0]-3), 5)
    elif (p == 'ma'):
        pos = (5, (im.shape[1]//2)-wm.shape[1]//2)
    elif (p == 'mm'):
        pos = ((im.shape[0]//2)-wm.shape[0]//2, (im.shape[1]//2)-wm.shape[1]//2)
    elif (p == 'md'):
        pos = ((im.shape[0]-wm.shape[0]-3), (im.shape[1]//2)-wm.shape[1]//2)
    elif (p == 'ra'):
        pos = (5, (im.shape[1]-wm.shape[1])-5)
    elif (p == 'rm'):
        pos = ((im.shape[0]//2)-wm.shape[0]//2, (im.shape[1]-wm.shape[1])-5)
    elif (p == 'rd'):
        pos = ((im.shape[0]-wm.shape[0])-3, (im.shape[1]-wm.shape[1])-5)
    return pos

#Take for arguments: a RGBA image, a RGBA watermark image, the position for the watermark and its transparency
#Adapt the transparency of the pixels in the given position
def transform_image(im, wm, pos, trns):
    wm  = resize_watermark(im, wm)
    p = convert_to_coordinates(im, wm, pos)
    for i in range(wm.shape[0]):
        for j in range(wm.shape[1]):
            if(wm[i,j][0]>0):
                im[p[0]+i, p[1]+j][3] = trns
    return im

#Take for arguments: an image, a watermark image, the position for the watermark, 
#its transparency and optionally if the user want to save it 
#Apply the watermark to the image at the given position with the given transparency
def apply_watermak(im, wm, pos, trns, file_name = None):
    im = transform_image(to_rgba(im), to_rgba(wm), pos, trns)
    if file_name:
        imsave('{}.png'.format(file_name), im)
    return im

#Take for arguments: an image, a watermark image, the position for the watermark
#and optionally if the user want to save it 
#Remove the watermark from the image at the given position
def remove_watermark(im, wm, pos, file_name = None):
    im = transform_image(to_rgba(im), to_rgba(wm), pos, 255)
    if file_name:
        imsave('{}.png'.format(file_name), im)
    return im

#fnt = ImageFont.truetype('/System/Library/Fonts/Supplemental/Example.ttf', 300) path for fonts for Mac users
#A sample of 6 fonts has been implemented together with the notebook
fnt = ImageFont.truetype('Arial.ttf', 300)
fnt1 = ImageFont.truetype('Chalkboard.ttc', 300)
fnt2 = ImageFont.truetype('Charter.ttc', 300)
fnt3 = ImageFont.truetype('Luminari.ttf', 300)
fnt4 = ImageFont.truetype('SnellRoundhand.ttc', 300)
fnt5 = ImageFont.truetype('Times New Roman.ttf', 300)
```

To create a watermark image use the create_im_watermark(...) function. 

There are 3 arguments: 
1. The text (type = str) which will be displayed in the watermark image. 
2. The font which will be used for the text.
3. (Optional) Determines if the watermark image created should be saved; default setting is False, use True to save the image. Example: wm = create_im_watermak("Text", fnt4, True)

```python
wm = create_im_watermak("Text", fnt4)
```

Apply the wanted watermark to your image with the apply_watermark(...) function.

There are five arguments:
1. The image
2. The watermark
3. The position either one of the 10 predetermined position or coordinates
4. The wanted transparency of the watermark
5. (Optional) Determines if the watermark image created should be saved and name it as the user has precised; default setting is False, indicate a name (type = str) to save it under this name. Example: im_wm_applied = apply_watermak(im, wm, "rd", 120, "Image_with_watermark")

```python
im_wm_applied = apply_watermak(im, wm, "rd", 120)

plt.figure(figsize=(15, 10))
plt.imshow(im_wm_applied)
plt.show()
```

Remove the watermark from your image with the remove_watermark(...) function.

There are four arguments:
1. The image
2. The watermark
3. The same position which was used to apply the watermark
4. (Optional) Determines if the watermark image created should be saved and name it as the user has precised; default setting is False, indicate a name (type = str) to save it under this name. Example: im_wm_removed = remove_watermark(im_wm_applied, wm, "rd", "Image_wm_removed")

```python
im_wm_removed = remove_watermark(im_wm_applied, wm, "rd")

plt.figure(figsize=(30, 22))
plt.subplot(2, 1, 1)
plt.imshow(im_wm_removed)
plt.title('Image with the Watermark removed, {}'.format(im_wm_removed.shape))
plt.subplot(2, 1, 2)
plt.imshow(im, cmap=plt.cm.gray, vmin=0, vmax=255)
plt.title('Original Image {}'.format(im.shape))
plt.show()
```
